```python
import numpy as np
import pandas as pd
excel_obj = pd.ExcelFile('isp_dirty_data.xlsx')

df = excel_obj.parse('Sheet1', dtype={'MSISDN': str})
display(df)
print('ตรวจสอบค่าว่าง รายคอลัม')
print(df.isna().sum())

df = df.dropna(subset=['MSISDN'])

df['RC_RATE'] = df['RC_RATE'].fillna(0)

df['CUST_NAME'] = df['CUST_NAME'].fillna('Unknown Customer')

df['CONTRACT_END_DT'] = pd.to_datetime(df['CONTRACT_END_DT'])

df['CONTRACT_END_DT'] = df['CONTRACT_END_DT'].fillna(pd.Timestamp('2099-12-31'))

display(df)
```

```python
df = pd.read_excel('../Lesson_2/isp_duplicate_data.xlsx', dtype={'MSISDN': str})

df['UPDATE_DATE'] = pd.to_datetime(df['UPDATE_DATE'])
display(df)
print('จำนวนแถวที่ซ้ำ')
print(df.duplicated(subset=['MSISDN']).sum())

df = df.sort_values('UPDATE_DATE', ascending=True)

df_clean = df.drop_duplicates(subset=['MSISDN'], keep='last')

print(f'จากเดิม {len(df)} แถว เหลือข้อมูลที่ถูกต้อง {len(df_clean)}')
display(df_clean)
```
