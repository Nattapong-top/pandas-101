7) โจทย์ฝึกเบื้องต้น (ไม่เฉลย)

เพื่อให้ป๋า “ร่อน list ได้ลื่นมาก”
ผมให้แนวคิดแค่สั้น ๆ ตามสไตล์ป๋า

7.1 หาค่าใน list ที่มากที่สุด

แนวคิด: เก็บ max ชั่วคราวแล้วไล่เทียบ


```python
nums = [1, 2, 3, 4]
total = None
for num in nums:
    if total is None or total < num:
        total = num
print(total)
```


7.2 แปลง list ให้เป็น list ใหม่ที่ทุกค่า +10

แนวคิด: list comprehension


```python
x = [n+10 for n in nums]
print(x)
```


7.3 ให้ matrix ซ้อน list แล้วหาผลรวมทั้ง matrix

แนวคิด: nested loop


```python
matrix = [
    [1, 2, 3, 4],
    [1, 2, 3, 4]
]
total = 0
for row in matrix:
    for n in row:
        total += n
total
```


7.4 หาผลรวมแต่ละแถว

แนวคิด: นับ row เป็น list


```python
result = []
for row in matrix:
    total = 0
    for n in row:
        total += n
    result.append(total)
result
```


7.5 หาผลรวมแต่ละคอลัมน์

แนวคิด: index-driven

ป๋าเคยทำ ColumnSum กับ ColumnMax แล้ว จำฟีลได้ชัดเลย!


```python
matrix = [
    [1, 2, 3, 4],
    [1, 2, 3, 4]
]
result = []
max_cols = max(len(m) for m in matrix)
result = [0] * max_cols


for row in matrix:
    for i in range(len(row)):
        result[i] += row[i]
result
```
