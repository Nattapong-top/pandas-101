3) เพิ่ม–ลบข้อมูล

ป๋าจะได้ใช้บ่อยมากเวลาจัดข้อมูลก่อนส่งเข้าตาราง

```python
lst = [1, 2, 3]

lst.append(4)   # ต่อท้าย
lst.insert(1, 99)       # แทรกที่ index 1
lst.remove(2)           # ลบค่า 2
del lst[0]              # ลบตาม index 0
print(lst)
```

4) Loop list แบบพื้นฐาน

หัวใจของ nested list/matrix เลย

```python
for x in lst:
    print(x)
```

แบบจับ index ด้วย:

```python
for i in range(len(lst)):
    print(i, lst[i])
```

5) List Comprehension

นี่เป็นสไตล์ที่ Pandas ทำงานคล้าย ๆ กัน เช่น map, filter

```python
nums = [1, 2, 3, 4]
sq = [x*x for x in nums]
evens = [x for x in nums if x%2==0]
```

6) Nested List (เข้าแบบลึกขึ้นนิดนึง)

คือ list ของ list

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]
```

เข้าถึง 1 ช่อง:

```python
matrix[1][-1]
print(matrix[1][-1])
```

เข้าถึงแบบ loop:

```python
for row in matrix:
    for col in row:
        print(col)
```

การเดินแถวแบบนี้ = พื้นฐานของการ Loop DataFrame เวลา iterate row เช่น .iterrows()
