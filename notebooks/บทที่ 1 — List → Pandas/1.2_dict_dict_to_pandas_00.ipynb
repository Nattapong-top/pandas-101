บทย่อย 1.2 — Dict พื้นฐาน (โครงสร้างข้อมูลแบบคีย์–ค่า)
1) Dict คืออะไร

เป็นโครงสร้างข้อมูลแบบ “key → value”
ใช้แทนข้อมูล 1 แถว ได้ดีที่สุด

```python
user = {
    'name': 'ป๋า',
    'age': 32,
    'dept': 'IT'
}
user
```

ประโยชน์:

อ่านง่ายเหมือนข้อมูลจริง

เข้าถึงด้วยชื่อคอลัมน์ ไม่ต้องจำ index

Pandas ใช้ dict เป็นฐานสำคัญสุด (list ของ dict)

```python
print(user['name'], user['age'], user['dept'])
```

ถ้าคีย์อาจไม่มี ใช้ .get() กันล้ม

```python
user.get('salary', 0)
# อันนี้ Pandas ก็มีฟีลคล้าย ๆ .get() เวลาบางคอลัมน์ไม่อยู่
```

3) เพิ่ม–แก้–ลบ key

```python
user['salary'] = 25000 # เพิ่ม
user['age'] = 36        # แก้
del user['dept']        # ลบ
user
```

4) Loop dict ใช้งานจริง

แบบ loop key:

```python
for k in user:
    print(k, user[k])
```

```python
# <!-- แบบโปร -->
for k, v in user.items():
    print(k, v)

    # ฟีลนี้เอาไปใช้กับ Pandas .iterrows() ได้เลย
```

5) Dict + List = โครงสร้างข้อมูลระดับโปร

โครงสร้างสำคัญที่สุดในงานจริง และใน Pandas:

5.1 List ของ dict → “เหมือน Excel แต่เป็น Python”

```python
rows = [
    {'name': 'A', 'age': 30},
    {'name': 'B', 'age': 25},
    {'name': 'C', 'age': 40}
]
rows
```

ป๋าเห็นไหม
แต่ละ dict = 1 row
list = ตาราง
คือ DataFrame แบบดิบ ๆ เลย!
