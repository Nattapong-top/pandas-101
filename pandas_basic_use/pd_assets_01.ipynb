'‡πÇ‡∏•‡∏∞‡∏Ñ‡∏≠‡∏°‡πÄ‡∏Å‡πà‡∏≤' ‡πÉ‡∏´‡πâ‡∏´‡∏ô‡πà‡∏≠‡∏¢ ‡πÄ‡∏≠‡∏≤‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏ó‡∏µ‡πà‡∏ã‡∏∑‡πâ‡∏≠ ‡∏Å‡πà‡∏≠‡∏ô‡∏õ‡∏µ 2021 ‡∏ó‡∏¥‡πâ‡∏á‡πÉ‡∏´‡πâ‡∏´‡∏°‡∏î ‡πÅ‡∏•‡πâ‡∏ß‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏á‡∏ö‡∏°‡∏≤‡∏ã‡∏¥‡∏ß‡πà‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ‡πÄ‡∏á‡∏¥‡∏ô‡∏ã‡∏∑‡πâ‡∏≠‡πÉ‡∏´‡∏°‡πà‡πÄ‡∏ó‡πà‡∏≤‡πÑ‡∏´‡∏£‡πà?"

‡∏ñ‡πâ‡∏≤‡∏õ‡πã‡∏≤‡πÉ‡∏ä‡πâ Excel ‡∏õ‡πã‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏≤‡∏ô‡∏±‡πà‡∏á‡∏Å‡∏£‡∏≠‡∏á Filter ‡∏ó‡∏µ‡∏•‡∏∞‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î... ‡πÅ‡∏ï‡πà‡∏ß‡∏±‡∏ô‡∏ô‡∏µ‡πâ‡∏õ‡πã‡∏≤‡∏Ñ‡∏∑‡∏≠ Programmer! ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏Ñ‡πâ‡∏î‡∏™‡∏±‡πà‡∏á‡∏ï‡∏≤‡∏¢‡∏Å‡∏±‡∏ô‡∏Ñ‡∏£‡∏±‡∏ö

```python
import pandas as pd
```

```python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≥‡∏•‡∏≠‡∏á (Asset IT)
data = {
    'Asset_ID': ['PC-001', 'PC-002', 'PC-003', 'PC-004', 'PC-005'],
    'Brand': ['Dell', 'HP', 'Lenovo', 'Dell', 'Acer'],
    'Year_Bought': [2018, 2022, 2019, 2023, 2017], # ‡∏õ‡∏µ‡∏ó‡∏µ‡πà‡∏ã‡∏∑‡πâ‡∏≠
    'RAM_GB': [4, 16, 8, 16, 4] # ‡πÅ‡∏£‡∏°
}
```

```python
df = pd.DataFrame(data)

print('--- ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏£‡∏±‡∏û‡∏¢‡πå‡∏™‡∏¥‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î ---')
print(df)
```

üéÆ ‡∏†‡∏≤‡∏£‡∏Å‡∏¥‡∏à‡∏ó‡∏µ‡πà 1: ‡∏ï‡∏≤‡∏°‡∏•‡πà‡∏≤‡∏´‡∏≤‡∏Ñ‡∏≠‡∏°‡πÄ‡∏Å‡πà‡∏≤ (Filter)
‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡πÄ‡∏à‡πâ‡∏≤‡∏ô‡∏≤‡∏¢‡∏™‡∏±‡πà‡∏á‡πÄ‡∏≠‡∏≤‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏ó‡∏µ‡πà‡∏ã‡∏∑‡πâ‡∏≠ ‡∏Å‡πà‡∏≠‡∏ô‡∏õ‡∏µ 2021 (< 2021) ‡πÉ‡∏ô Pandas ‡πÄ‡∏£‡∏≤‡πÉ‡∏ä‡πâ‡∏Å‡∏≤‡∏£ "‡πÉ‡∏™‡πà‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡πÉ‡∏ô‡∏ß‡∏á‡πÄ‡∏•‡πá‡∏ö‡πÄ‡∏´‡∏•‡∏µ‡πà‡∏¢‡∏°" ‡∏Ñ‡∏£‡∏±‡∏ö

```python
# ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô df ‡∏ó‡∏µ‡πà‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå Year_Bought ‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ 2021
old_computers = df[df['Year_Bought'] < 2021]

print('--- ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≠‡∏°‡πÄ‡∏Å‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÇ‡∏•‡∏∞ ---')
print(old_computers)
```

üéÆ ‡∏†‡∏≤‡∏£‡∏Å‡∏¥‡∏à‡∏ó‡∏µ‡πà 2: ‡∏Ç‡∏≠‡∏á‡∏ö‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì (Calculation)
‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡πÄ‡∏à‡πâ‡∏≤‡∏ô‡∏≤‡∏¢‡∏ñ‡∏≤‡∏°‡∏ß‡πà‡∏≤ "‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ‡πÄ‡∏á‡∏¥‡∏ô‡πÄ‡∏ó‡πà‡∏≤‡πÑ‡∏´‡∏£‡πà?" ‡∏™‡∏°‡∏°‡∏ï‡∏¥‡∏ß‡πà‡∏≤‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏ã‡∏∑‡πâ‡∏≠‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡πÉ‡∏´‡∏°‡πà ‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏•‡∏∞ 25,000 ‡∏ö‡∏≤‡∏ó

‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡πÄ‡∏Å‡πà‡∏≤ ‡πÅ‡∏•‡πâ‡∏ß‡∏Ñ‡∏π‡∏ì 25,000 ‡∏Ñ‡∏£‡∏±‡∏ö

```python
# 1. ‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏ó‡∏µ‡πà‡∏Å‡∏£‡∏≠‡∏á‡∏°‡∏≤‡πÑ‡∏î‡πâ (‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á len ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô len(list))
count = len(old_computers)

# 2. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡πÄ‡∏á‡∏¥‡∏ô
total_budget = count * 25000

print(f'\n‡∏™‡∏£‡∏∏‡∏õ: ‡∏°‡∏µ‡∏Ñ‡∏≠‡∏°‡πÄ‡∏Å‡πà‡∏≤ {count} ‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á')
print(f'‡∏ï‡πâ‡∏≠‡∏á‡∏Ç‡∏≠‡∏á‡∏ö‡∏õ‡∏£‡∏∞‡∏°‡∏≤‡∏ì‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {total_budget:,} ‡∏ö‡∏≤‡∏ó')
```

üöÄ ‡∏ï‡∏≤‡∏ô‡∏µ‡πâ‡∏õ‡πã‡∏≤‡πÄ‡∏•‡πà‡∏ô‡∏ö‡πâ‡∏≤‡∏á (‡πÅ‡∏ö‡∏ö‡∏ù‡∏∂‡∏Å‡∏´‡∏±‡∏î)
‡∏à‡∏∏‡∏Å‡πÄ‡∏á‡∏¥‡∏ô‡∏Ç‡∏≠‡∏ó‡πâ‡∏≤‡∏õ‡πã‡∏≤‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏ô‡∏¥‡∏î‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Ñ‡∏£‡∏±‡∏ö...

‡∏ñ‡πâ‡∏≤‡πÄ‡∏à‡πâ‡∏≤‡∏ô‡∏≤‡∏¢‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÉ‡∏à ‡∏ö‡∏≠‡∏Å‡∏ß‡πà‡∏≤ "‡πÑ‡∏°‡πà‡∏î‡∏π‡∏õ‡∏µ‡πÅ‡∏•‡πâ‡∏ß... ‡πÅ‡∏ï‡πà‡∏à‡∏∞‡πÇ‡∏•‡∏∞‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏ó‡∏µ‡πà ‡πÅ‡∏£‡∏°‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ 8 GB ‡∏≠‡∏≠‡∏Å‡πÉ‡∏´‡πâ‡∏´‡∏°‡∏î"

```python
com_ram_min = df[df['RAM_GB'] < 8]
print('--- ‡∏Ñ‡∏≠‡∏°‡πÅ‡∏£‡∏°‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ 8GB ---')
print(com_ram_min)

count_ram_min = len(com_ram_min)

budget_com_new = count_ram_min * 25000
print(budget_com_new)
```

```python
# ‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏ï‡∏≤‡∏£‡∏≤‡∏á com_ram_min ‡∏•‡∏á‡πÑ‡∏ü‡∏•‡πå‡∏ä‡∏∑‡πà‡∏≠ "request_buy_new.xlsx"
# index=False ‡∏Ñ‡∏∑‡∏≠ ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏≠‡∏≤‡πÄ‡∏•‡∏Ç‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î 0, 1, 2 ‡∏ï‡∏¥‡∏î‡πÑ‡∏õ‡∏î‡πâ‡∏ß‡∏¢ (‡πÄ‡∏à‡πâ‡∏≤‡∏ô‡∏≤‡∏¢‡πÑ‡∏°‡πà‡∏≠‡∏¢‡∏≤‡∏Å‡∏î‡∏π)
# com_ram_min.to_excel('../output/request_buy_new.xlsx', index=False)
old_computers.to_excel('../output/request_buy_new.xlsx', index=False)
print('‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡πÑ‡∏ü‡∏•‡πå‡πÄ‡∏™‡∏£‡πá‡∏à‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢!')
```
