```python
import pandas as pd

# 1. อ่านไฟล์ (เหมือนเปิดไฟล์ excel)
df = pd.read_csv('../datasets/stock.csv')
df.head()

```

```python
# 2. สร้างคอลัมน์ใหม่ "กำไรทั้งหมด" (Total_Profit)
# สูตร: (ราคาขาย - ราคาทุน) * จำนวนที่ขาย
df['Total_Profit'] = (df['Price'] - df['Cost']) * df['Quantity']
total = df['Total_Profit'].sum()
# 3. โชว์ผลลัพธ์
print(df)
print(total)
```
