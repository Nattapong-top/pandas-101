```python
import pandas as pd
import numpy as np

data_new = {
    'S_Number': ['TH-888', 'th-888', 'TH-999', 'TH-111', 'th-222', 'TH-222'],
    'Vender': [' asus ', 'Asus', ' Apple ', 'apple', 'Microsoft', 'MICROSOFT'],
    'Spec_RAM': ['8GB', '8 GB', '16384MB', '4GB', '16GB', '16384MB'],
    'Price_Tag': ['‡∏ø35,000', '35000', '$1,200', '‡∏ø28,500', '42,000', '‡∏ø42000'],
    'Receive_Date': ['2025.12.30', '2025/12/30', '05-01-2026', '2025.11.15', '10/01/2026', '2026.01.10']
}

df_store_raw = pd.DataFrame(data_new)
print("--- ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏î‡∏¥‡∏ö‡∏•‡πá‡∏≠‡∏ï‡πÉ‡∏´‡∏°‡πà ---")
print(df_store_raw)
```

üõ†Ô∏è ‡∏†‡∏≤‡∏£‡∏Å‡∏¥‡∏à‡∏Ç‡∏≠‡∏á‡∏õ‡πã‡∏≤ (6 ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏™‡∏π‡πà‡∏Ñ‡∏ß‡∏≤‡∏°‡∏Ñ‡∏•‡πà‡∏≠‡∏á) ‡∏õ‡πã‡∏≤‡∏•‡∏≠‡∏á‡πÉ‡∏ä‡πâ‡∏ó‡πà‡∏≤‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏°‡∏≤ ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£ df_store ‡∏ï‡∏≤‡∏°‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏ô‡∏µ‡πâ‡∏ô‡∏∞‡∏Ñ‡∏£‡∏±‡∏ö:
‡∏•‡πâ‡∏≤‡∏á‡∏ä‡∏∑‡πà‡∏≠‡∏¢‡∏µ‡πà‡∏´‡πâ‡∏≠ (Vender): ‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡∏î‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡∏ó‡∏¥‡πâ‡∏á


```python
df_store = df_store_raw.copy()
df_store['Vender'] = df_store['Vender'].str.strip().str.upper()
df_store
```


‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£ RAM (Spec_RAM): ‡πÅ‡∏õ‡∏•‡∏á 16384MB ‡πÉ‡∏´‡πâ‡∏Å‡∏•‡∏≤‡∏¢‡πÄ‡∏õ‡πá‡∏ô 16 (GB) ‡πÅ‡∏•‡∏∞‡∏•‡∏ö‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏ó‡∏¥‡πâ‡∏á‡πÉ‡∏´‡πâ‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÅ‡∏ï‡πà‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (‡πÉ‡∏ä‡πâ‡∏ó‡πà‡∏≤ Pythonic ‡∏´‡∏£‡∏∑‡∏≠ Lambda ‡∏ó‡∏µ‡πà‡∏õ‡πã‡∏≤‡∏ä‡∏≠‡∏ö‡∏Å‡πá‡πÑ‡∏î‡πâ‡∏Ñ‡∏£‡∏±‡∏ö)


```python

nums = df_store['Spec_RAM'].str.extract(r'(\d+)')[0].astype(float)

df_store['Spec_RAM'] = np.where(
    df_store['Spec_RAM'].str.contains('MB', case=False),
    nums / 1024,
    nums
).astype(int)
print(df_store)
```


‡∏•‡πâ‡∏≤‡∏á‡∏£‡∏≤‡∏Ñ‡∏≤ (Price_Tag): ‡∏•‡∏ö‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢ ‡∏ø, $, ‡πÅ‡∏•‡∏∞ , ‡∏≠‡∏≠‡∏Å‡πÉ‡∏´‡πâ‡∏´‡∏°‡∏î 

```python
df_store['Price_Tag'] = df_store['Price_Tag'].replace(r'[^0-9.]', '', regex=True).astype(float)
df_store
```

‡πÅ‡∏•‡πâ‡∏ß‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç float‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà (Receive_Date): ‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏õ‡∏µ-‡πÄ‡∏î‡∏∑‡∏≠‡∏ô-‡∏ß‡∏±‡∏ô (YYYY-MM-DD) ‡πÉ‡∏´‡πâ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏Å‡∏±‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î


```python
df_store['Receive_Date'] = pd.to_datetime(
    df_store['Receive_Date'],
    dayfirst=True,
    format='mixed',
    errors='coerce'
)
df_store['Receive_Date'] = df_store['Receive_Date'].dt.strftime('%Y-%m-%d')
df_store
```


‡∏•‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡πâ‡∏≥: ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö S_Number (‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏°‡∏ó‡∏≥‡∏ï‡∏±‡∏ß‡πÉ‡∏´‡∏ç‡πà‡∏Å‡πà‡∏≠‡∏ô) ‡πÅ‡∏•‡πâ‡∏ß‡∏•‡∏ö‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏ã‡πâ‡∏≥‡∏≠‡∏≠‡∏Å


```python
df_store['S_Number'] = df_store['S_Number'].str.upper()
df_store = df_store.drop_duplicates(subset=['S_Number'])
df_store
```


‡∏™‡∏£‡∏∏‡∏õ‡∏¢‡∏≠‡∏î: ‡∏´‡∏≤‡∏ú‡∏•‡∏£‡∏ß‡∏°‡∏£‡∏≤‡∏Ñ‡∏≤‡πÅ‡∏¢‡∏Å‡∏ï‡∏≤‡∏°‡∏¢‡∏µ‡πà‡∏´‡πâ‡∏≠ (Vender)


```python
sum_group = df_store.groupby('Vender')['Price_Tag'].sum()

for no, vander in enumerate(sum_group.items()) :
    brand, price = vander
    print(f'{no+1} ‡∏¢‡∏≠‡∏î‡∏£‡∏ß‡∏°‡∏¢‡∏µ‡πà‡∏´‡πâ‡∏≠: {brand:<10} ‡∏£‡∏≤‡∏Ñ‡∏≤: {price:,.2f} ‡∏ö‡∏≤‡∏ó')
```

3: ‡πÉ‡∏ä‡πâ Regex r'[^0-9.]' ‡∏à‡∏∞‡∏ä‡πà‡∏ß‡∏¢‡∏Å‡∏ß‡∏≤‡∏î‡∏Ç‡∏¢‡∏∞‡πÉ‡∏ô‡∏£‡∏≤‡∏Ñ‡∏≤‡πÑ‡∏î‡πâ‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡∏Ñ‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠
