‡∏≠‡∏ö‡∏ô‡∏µ‡πâ‡∏ú‡∏°‡∏à‡∏±‡∏î‡∏ä‡∏∏‡∏î "IT Inventory Management" (‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏û‡∏±‡∏™‡∏î‡∏∏‡πÑ‡∏≠‡∏ó‡∏µ) ‡∏°‡∏≤‡πÉ‡∏´‡πâ‡∏Ñ‡∏£‡∏±‡∏ö ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ô‡∏µ‡πâ‡∏à‡∏∞‡∏à‡∏≥‡∏•‡∏≠‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≠‡∏°‡∏û‡∏¥‡∏ß‡πÄ‡∏ï‡∏≠‡∏£‡πå‡πÉ‡∏ô‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏° "‡∏°‡∏±‡πà‡∏ß" ‡∏™‡∏π‡∏á‡∏Å‡∏ß‡πà‡∏≤‡πÄ‡∏î‡∏¥‡∏° ‡∏ó‡∏±‡πâ‡∏á‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏ß‡∏±‡∏î (GB/MB), ‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏á‡∏¥‡∏ô‡∏ï‡∏£‡∏≤‡∏´‡∏•‡∏≤‡∏¢‡∏™‡∏Å‡∏∏‡∏• ‡πÅ‡∏•‡∏∞‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏ó‡∏µ‡πà‡∏û‡∏¥‡∏°‡∏û‡πå‡∏°‡∏≤‡∏Ñ‡∏ô‡∏•‡∏∞‡∏ü‡∏≠‡∏£‡πå‡πÅ‡∏°‡∏ï‡∏Ñ‡∏£‡∏±‡∏ö

üõ†Ô∏è ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ó‡∏µ‡πà 1: ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏ä‡∏∏‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏û‡∏±‡∏™‡∏î‡∏∏‡πÑ‡∏≠‡∏ó‡∏µ (Messy Inventory)
‡∏Å‡πä‡∏≠‡∏õ‡∏õ‡∏µ‡πâ‡πÇ‡∏Ñ‡πâ‡∏î‡∏ô‡∏µ‡πâ‡πÑ‡∏õ‡∏£‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏à‡∏ó‡∏¢‡πå‡∏Ñ‡∏£‡∏±‡∏ö

```python
import pandas as pd
import numpy as np

inv_data = {
    'Serial_Number': ['SN-001', 'sn-001', 'SN-002', 'SN-003', 'SN-004', 'SN-005', 'SN-006'],
    'Brand': ['dell', 'DELL', 'hp', '  Apple  ', 'Lenovo', 'hp', 'DELL'],
    'Model': ['Latitude 5420', 'Latitude 5420', 'EliteBook', '  MacBook Pro  ', 'ThinkPad T14', 'EliteBook', 'Latitude 7420'],
    'RAM_Size': ['16GB', '16GB', '8GB', '16GB', '32GB', '4096MB', '16GB'],
    'Purchase_Price': ['‡∏ø25,000', '25000', '$800', '‡∏ø55,000', '32,000', '‡∏ø22000', '28000'],
    'Purchase_Date': ['2024.01.15', '2024/01/15', '12-05-2023', '2024.02.20', '2023.11.10', '2023/12/01', '2024.01.20'],
    'Status': ['In Use', 'In Use', 'Available', 'Repair', 'In-Use', 'Available', 'IN USE']
}

df_inv_origi = pd.DataFrame(inv_data)
df_inv = df_inv_origi.copy()
print("‡∏£‡∏∞‡∏ö‡∏ö‡∏Ñ‡∏•‡∏±‡∏á‡∏û‡∏±‡∏™‡∏î‡∏∏‡πÑ‡∏≠‡∏ó‡∏µ (‡∏â‡∏ö‡∏±‡∏ö‡∏ï‡πâ‡∏≠‡∏á‡∏ä‡∏≥‡∏£‡∏∞‡∏•‡πâ‡∏≤‡∏á) ‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÅ‡∏•‡πâ‡∏ß‡∏Ñ‡∏£‡∏±‡∏ö‡∏õ‡πã‡∏≤!")
print(df_inv)
```

‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå Brand ‡πÅ‡∏•‡∏∞ Model ‡∏°‡∏µ‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏•‡πá‡∏Å-‡πÉ‡∏´‡∏ç‡πà‡∏õ‡∏ô‡∏Å‡∏±‡∏ô ‡∏à‡∏á‡∏•‡∏ö‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏õ‡∏£‡∏±‡∏ö‡πÄ‡∏õ‡πá‡∏ô ‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î (Upper Case)

```python
'''‡πÅ‡∏ö‡∏ö‡∏á‡πà‡∏≤‡∏¢'''
df_invv = df_inv.copy()
df_invv['Brand'] = df_inv['Brand'].str.strip().str.upper()
df_invv['Model'] = df_inv['Model'].str.strip().str.upper()
df_invv

```

```python
'''‡πÅ‡∏ö‡∏ö‡∏°‡∏∑‡∏≠‡∏≠‡∏≤‡∏ä‡∏µ‡∏û'''
# df_inv2 = df_inv.copy()
cols_to_clean = ['Brand', 'Model']
df_inv[cols_to_clean] = df_inv[cols_to_clean].apply(lambda X: X.str.strip().str.upper())
df_inv
```

```python
'''‡πÅ‡∏ö‡∏ö‡∏ò‡∏£‡∏£‡∏°‡∏î‡∏≤'''
df_inv3 = df_inv.copy()
for col in cols_to_clean:
    df_inv3[col] = df_inv3[col].str.strip().str.upper()
df_inv3
```


‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡πÉ‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå RAM_Size ‡πÉ‡∏´‡πâ‡∏•‡∏ö‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£ 'GB' ‡∏≠‡∏≠‡∏Å‡πÉ‡∏´‡πâ‡∏´‡∏°‡∏î


```python
'''‡πÅ‡∏ö‡∏ö‡∏á‡πà‡∏≤‡∏¢‡πÜ'''
df_inv_ram = df_inv.copy()
df_inv_ram['RAM_Size'] = df_inv_ram['RAM_Size'].replace(r'[GB]', '', regex=True)

df_inv_ram
```

```python
'''‡πÅ‡∏ö‡∏ö‡∏á‡πà‡∏≤‡∏¢‡πÜ‡∏≠‡∏µ‡∏Å‡∏ß‡∏¥‡∏ò‡∏µ'''
df_inv_ram2 = df_inv.copy()
clean_ram = {
    '16GB': '16',
    '8GB': '8',
    '32GB': '32',
    '4096MB': '4'
}
df_inv_ram2['RAM_Size'] = df_inv_ram2['RAM_Size'].replace(clean_ram).astype(int)
df_inv_ram2
```

```python
''''''
```


‡πÇ‡∏à‡∏ó‡∏¢‡πå: (‡∏¢‡∏≤‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô!) ‡πÉ‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå RAM_Size ‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô '4096MB' ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÄ‡∏õ‡πá‡∏ô '4' (‡πÄ‡∏û‡∏£‡∏≤‡∏∞ 1024MB = 1GB)

‡∏Ñ‡∏≥‡πÉ‡∏ö‡πâ: df_inv['RAM_Size'].replace('4096MB', '4', inplace=True)


```python
'''‡πÅ‡∏ö‡∏ö‡∏ò‡∏£‡∏£‡∏°‡∏î‡∏≤'''
df_inv4 = df_inv.copy()
df_inv4['RAM_Size'] = df_inv4['RAM_Size'].replace(['4096MB'], ['4'])
df_inv4
```

```python
'''‡πÅ‡∏ö‡∏ö‡∏°‡∏∑‡∏≠‡∏≠‡∏≤‡∏ä‡∏µ‡∏û'''
# 1. ‡∏î‡∏∂‡∏á‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡πÅ‡∏•‡πâ‡∏ß‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô float
# (\d+) ‡∏Ñ‡∏∑‡∏≠‡πÄ‡∏≠‡∏≤‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
df_inv_ram3 = df_inv.copy()
nums = df_inv_ram3['RAM_Size'].str.extract(r'(\d+)')[0].astype(float)

# 2. ‡πÉ‡∏ä‡πâ np.where (‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏™‡∏π‡∏ï‡∏£ IF ‡πÉ‡∏ô Excel)
# ‡∏ñ‡πâ‡∏≤‡πÉ‡∏ô RAM_Size ‡∏°‡∏µ‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤ 'MB' ‡πÉ‡∏´‡πâ‡πÄ‡∏≠‡∏≤‡πÄ‡∏•‡∏Ç‡πÑ‡∏õ‡∏´‡∏≤‡∏£ 1024 ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ (‡πÄ‡∏õ‡πá‡∏ô GB) ‡πÉ‡∏´‡πâ‡πÉ‡∏ä‡πâ‡πÄ‡∏•‡∏Ç‡πÄ‡∏î‡∏¥‡∏°‡πÄ‡∏•‡∏¢
df_inv_ram3['RAM_Size'] = np.where(
    df_inv['RAM_Size'].str.contains('MB', case=False),
    nums / 1024,
    nums
).astype(int)
print(df_inv_ram3['RAM_Size'])
```

```python
'''‡πÅ‡∏ö‡∏ö‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à‡∏á‡πà‡∏≤‡∏¢'''
def convert_ram(value):
    value = str(value).upper().strip()

    if 'MB' in value:
        number = float(value.replace('MB', ''))
        return int(number / 2024)
    elif 'GB' in value:
        return int(value.replace('GB', ''))
    return value

df_inv['RAM_Size'] = df_inv['RAM_Size'].apply(convert_ram)

df_inv['RAM_Size'].info()
```

‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏õ‡∏£‡∏±‡∏ö‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡πÉ‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå Status ‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô 'In-Use' ‡∏´‡∏£‡∏∑‡∏≠ 'IN USE' ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤ 'In Use' ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î


```python
df_inv['Status'] = df_inv['Status'].replace(['In-Use', 'IN USE'], 'In Use')
df_inv
```


‡∏´‡∏°‡∏ß‡∏î: ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏á‡πÄ‡∏á‡∏¥‡∏ô‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (Currency & Numeric) 

5. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå Purchase_Price ‡∏°‡∏µ‡∏ó‡∏±‡πâ‡∏á‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢ ‡∏ø, $, ‡πÅ‡∏•‡∏∞ , ‡∏à‡∏á‡∏•‡∏ö‡∏™‡∏±‡∏ç‡∏•‡∏±‡∏Å‡∏©‡∏ì‡πå‡πÄ‡∏´‡∏•‡πà‡∏≤‡∏ô‡∏µ‡πâ‡∏≠‡∏≠‡∏Å‡πÉ‡∏´‡πâ‡∏´‡∏°‡∏î (‡πÉ‡∏ä‡πâ Regex ‡∏ó‡πà‡∏≤‡πÑ‡∏°‡πâ‡∏ï‡∏≤‡∏¢‡∏ó‡∏µ‡πà‡∏õ‡πã‡∏≤‡πÄ‡∏û‡∏¥‡πà‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏°‡∏≤‡πÑ‡∏î‡πâ‡πÄ‡∏•‡∏¢!) 


```python
df_inv['Purchase_Price'] = df_inv['Purchase_Price'].replace(r'[‡∏ø$, ]', '', regex=True)
df_inv
```


6. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå Purchase_Price ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (Float) ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏£‡∏≠‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì


```python
df_inv['Purchase_Price'] = df_inv['Purchase_Price'].astype(float)
df_inv
```

‡∏´‡∏°‡∏ß‡∏î: ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÅ‡∏•‡∏∞‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡πâ‡∏≥ (DateTime & Duplicates) 

7. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏•‡∏ö‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà Serial_Number ‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô‡∏≠‡∏≠‡∏Å (‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏°‡∏õ‡∏£‡∏±‡∏ö Serial_Number ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏ä‡πá‡∏Ñ‡∏ã‡πâ‡∏≥‡∏ô‡∏∞‡∏Ñ‡∏£‡∏±‡∏ö‡∏õ‡πã‡∏≤!) 


```python
df_inv['Serial_Number'] = df_inv['Serial_Number'].str.upper()
df_inv = df_inv.drop_duplicates(subset=['Serial_Number'])
df_inv
```


8. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: (‡∏ó‡πâ‡∏≤‡∏ó‡∏≤‡∏¢!) ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå Purchase_Date ‡∏°‡∏µ‡∏ó‡∏±‡πâ‡∏á‡∏à‡∏∏‡∏î (.) ‡πÅ‡∏•‡∏∞‡∏™‡πÅ‡∏•‡∏ä (/) ‡∏à‡∏á‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢‡∏Ç‡∏µ‡∏î (-) ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î


```python
'''‡πÅ‡∏ö‡∏ö‡∏ò‡∏£‡∏£‡∏°‡∏î‡∏≤'''
df_inv_date = df_inv.copy()
df_inv_date['Purchase_Date'] = df_inv['Purchase_Date'].str.replace(r'[./]', '-', regex=True)
df_inv_date
```

```python
'''‡πÅ‡∏ö‡∏ö‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏à‡∏£‡∏¥‡∏á ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏°‡∏±‡∏ô‡∏°‡∏µ‡∏´‡∏•‡∏≤‡∏¢‡πÅ‡∏ö‡∏ö (‡∏ó‡πà‡∏≤‡∏ô‡∏µ‡πâ ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏à‡∏ö)'''
# ‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á‡πÄ‡∏ó‡∏û‡∏ï‡∏±‡∏ß‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏à‡∏ö
# errors='coerce' ‡∏Ñ‡∏∑‡∏≠‡∏ñ‡πâ‡∏≤‡πÄ‡∏à‡∏≠‡∏≠‡∏±‡∏ô‡πÑ‡∏´‡∏ô‡πÄ‡∏ô‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á‡πÜ ‡πÉ‡∏´‡πâ‡∏õ‡∏•‡πà‡∏≠‡∏¢‡∏ß‡πà‡∏≤‡∏á (NaT) ‡∏≠‡∏¢‡πà‡∏≤‡πÄ‡∏û‡∏¥‡πà‡∏á‡∏û‡πà‡∏ô Error
# dayfirst=True ‡∏ö‡∏≠‡∏Å‡∏ß‡πà‡∏≤ "‡∏ñ‡πâ‡∏≤‡πÄ‡∏à‡∏≠‡πÄ‡∏•‡∏Ç‡∏ó‡∏µ‡πà‡∏Å‡∏≥‡∏Å‡∏ß‡∏° (‡πÄ‡∏ä‡πà‡∏ô 12-05) ‡πÉ‡∏´‡πâ‡∏Ñ‡∏¥‡∏î‡∏ß‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô ‡∏ß‡∏±‡∏ô-‡πÄ‡∏î‡∏∑‡∏≠‡∏ô ‡πÑ‡∏ß‡πâ‡∏Å‡πà‡∏≠‡∏ô‡∏ô‡∏∞"
df_inv_date2 = df_inv.copy()
df_inv_date2['Purchase_Date'] = pd.to_datetime(df_inv_date2['Purchase_Date'], dayfirst=True, errors='coerce')
df_inv_date2
```

```python
'''‡πÅ‡∏ö‡∏ö‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏à‡∏£‡∏¥‡∏á ‡∏Ç‡∏±‡πâ‡∏ô‡∏™‡∏∏‡∏î'''
df_inv = df_inv.copy()
df_inv['Purchase_Date'] = pd.to_datetime(
    df_inv['Purchase_Date'],
    dayfirst=True,      # ‡∏ö‡∏≠‡∏Å‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤ "‡∏ñ‡πâ‡∏≤‡πÄ‡∏à‡∏≠‡πÄ‡∏•‡∏Ç‡∏Å‡∏≥‡∏Å‡∏ß‡∏° ‡πÉ‡∏´‡πâ‡πÄ‡∏≠‡∏≤ ‡∏ß‡∏±‡∏ô ‡∏Ç‡∏∂‡πâ‡∏ô‡∏Å‡πà‡∏≠‡∏ô‡∏ô‡∏∞"
    format='mixed',     # ‡πÑ‡∏°‡πâ‡∏ï‡∏≤‡∏¢: ‡∏ö‡∏≠‡∏Å‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤ "‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏°‡∏±‡∏ô‡∏ú‡∏™‡∏°‡∏Å‡∏±‡∏ô‡∏°‡∏≤ ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡πÉ‡∏´‡πâ‡∏´‡∏ô‡πà‡∏≠‡∏¢!"
    errors='coerce'     # ‡∏ñ‡πâ‡∏≤‡∏≠‡∏±‡∏ô‡πÑ‡∏´‡∏ô‡πÄ‡∏ô‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á‡πÜ (‡πÄ‡∏ä‡πà‡∏ô 31/02/2024) ‡πÉ‡∏´‡πâ‡∏Ç‡∏∂‡πâ‡∏ô NaT ‡πÑ‡∏ß‡πâ‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏à‡∏∞‡πÑ‡∏î‡πâ‡πÑ‡∏°‡πà‡∏û‡∏±‡∏á
)
df_inv['Purchase_Date'] = df_inv['Purchase_Date'].dt.strftime('%Y-%m-%d')
df_inv
```


‡∏´‡∏°‡∏ß‡∏î: ‡∏™‡∏£‡∏∏‡∏õ‡∏ú‡∏•‡∏£‡∏≤‡∏¢‡∏á‡∏≤‡∏ô (Final Report) 

9. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡∏°‡πà‡∏ä‡∏∑‡πà‡∏≠ Is_High_End ‡πÇ‡∏î‡∏¢‡∏ñ‡πâ‡∏≤ Purchase_Price ‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ 30,000 ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô True ‡∏ô‡∏≠‡∏Å‡∏ô‡∏±‡πâ‡∏ô‡πÄ‡∏õ‡πá‡∏ô False
