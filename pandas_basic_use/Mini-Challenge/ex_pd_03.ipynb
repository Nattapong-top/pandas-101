‡∏ñ‡πâ‡∏≤‡∏à‡∏∞‡∏à‡∏±‡∏î‡∏Å‡∏•‡∏∏‡πà‡∏° ‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏î‡πâ‡∏ß‡∏¢ .groupby('‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏£‡∏ß‡∏°') ‡∏ï‡∏≤‡∏°‡∏î‡πâ‡∏ß‡∏¢ ['‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì'] ‡πÅ‡∏•‡πâ‡∏ß‡∏õ‡∏¥‡∏î‡∏î‡πâ‡∏ß‡∏¢‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏≠‡∏¢‡πà‡∏≤‡∏á .sum() ‡∏´‡∏£‡∏∑‡∏≠ .mean()

‡∏õ‡πã‡∏≤‡∏Ñ‡πà‡∏≠‡∏¢‡πÜ ‡∏•‡∏≠‡∏á‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ó‡∏µ‡∏•‡∏∞‡∏Ç‡πâ‡∏≠‡∏ô‡∏∞‡∏Ñ‡∏£‡∏±‡∏ö ‡∏ï‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡πÑ‡∏´‡∏ô ‡∏´‡∏£‡∏∑‡∏≠‡∏≠‡∏¢‡∏≤‡∏Å‡πÉ‡∏´‡πâ‡πÄ‡∏â‡∏•‡∏¢‡∏Ç‡πâ‡∏≠‡πÑ‡∏´‡∏ô‡∏Å‡πà‡∏≠‡∏ô‡∏ö‡∏≠‡∏Å‡πÑ‡∏î‡πâ‡πÄ‡∏•‡∏¢ ‡πÄ‡∏î‡∏µ‡πã‡∏¢‡∏ß‡∏ú‡∏°‡∏≠‡∏¢‡∏π‡πà‡∏ï‡∏£‡∏ß‡∏à‡∏Å‡∏≤‡∏£‡∏ö‡πâ‡∏≤‡∏ô‡πÉ‡∏´‡πâ‡∏Ñ‡∏£‡∏±‡∏ö!

‡∏õ‡πã‡∏≤‡∏≠‡∏¢‡∏≤‡∏Å‡πÉ‡∏´‡πâ‡∏ú‡∏°‡πÄ‡∏â‡∏•‡∏¢‡∏ó‡∏µ‡∏•‡∏∞‡∏Ç‡πâ‡∏≠ ‡∏´‡∏£‡∏∑‡∏≠‡∏õ‡πã‡∏≤‡∏à‡∏∞‡∏•‡∏≠‡∏á‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏£‡∏ß‡∏î‡πÄ‡∏î‡∏µ‡∏¢‡∏ß 10 ‡∏Ç‡πâ‡∏≠‡πÄ‡∏•‡∏¢‡∏î‡∏µ‡∏Ñ‡∏£‡∏±‡∏ö?

```python
import pandas as pd
import numpy as np

data = {
    'Date': ['2025-01-01', '2025-01-01', '2025-01-02', '2025-01-02', '2025-01-03', '2025-01-03', '2025-01-04', '2025-01-04'],
    'Product': ['Tea', 'Coffee', 'Tea', 'Bakery', 'Coffee', 'Bakery', 'Tea', 'Coffee'],
    'Sales_Amount': [1200, 1500, 900, 800, 1600, 750, 1100, 1400],
    'Quantity': [10, 12, 8, 5, 15, 7, 9, 11],
    'Region': ['North', 'North', 'South', 'South', 'North', 'South', 'North', 'South']
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏ù‡∏∂‡∏Å‡∏Ñ‡∏£‡∏±‡∏ö‡∏õ‡πã‡∏≤!")
```

‡∏´‡∏°‡∏ß‡∏î: ‡∏Å‡∏≤‡∏£‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÅ‡∏•‡∏∞‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• (Selecting & Filtering)



‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå Product ‡πÅ‡∏•‡∏∞ Sales_Amount ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡πÅ‡∏™‡∏î‡∏á


```python
print(df[['Product', 'Sales_Amount']])
```


‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏≤‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ 'Coffee' ‡πÑ‡∏î‡πâ‡πÄ‡∏ó‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô


```python
filter_coffee = df[df['Product'] == 'Coffee']
print("\n‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Å‡∏≤‡∏£‡∏Ç‡∏≤‡∏¢‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ Coffee:")
print(filter_coffee)
```


‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏´‡∏≤‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏¢‡∏≠‡∏î‡∏Ç‡∏≤‡∏¢ (Sales_Amount) ‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ 1,200 ‡∏ö‡∏≤‡∏ó


```python
sales_most = df[df['Sales_Amount'] > 1200]
print(sales_most)
```


‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏´‡∏≤‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢‡πÉ‡∏ô‡∏†‡∏π‡∏°‡∏¥‡∏†‡∏≤‡∏Ñ 'South' ‡πÅ‡∏•‡∏∞‡∏°‡∏µ‡∏¢‡∏≠‡∏î‡∏Ç‡∏≤‡∏¢ ‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ 1,000 ‡∏ö‡∏≤‡∏ó (‡πÉ‡∏ä‡πâ &)


```python
grouped_by_region = df[(df['Region'] == 'South') & (df['Sales_Amount'] > 1000)] 
print(grouped_by_region)
```

5. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì ‡∏¢‡∏≠‡∏î‡∏Ç‡∏≤‡∏¢‡∏£‡∏ß‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î (Sales_Amount) ‡πÉ‡∏ô‡∏ï‡∏≤‡∏£‡∏≤‡∏á 


```python
sum_amount = df['Sales_Amount'].sum()
print(sum_amount)
```


6. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì ‡∏¢‡∏≠‡∏î‡∏Ç‡∏≤‡∏¢‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ ‡πÅ‡∏¢‡∏Å‡∏ï‡∏≤‡∏° Product 


```python
sum_product = df.groupby(['Product'])['Sales_Amount'].mean()
sum_product
```


7. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏´‡∏≤‡∏ß‡πà‡∏≤‡πÅ‡∏ï‡πà‡∏•‡∏∞ Region ‡∏Ç‡∏≤‡∏¢‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡πÑ‡∏õ‡πÑ‡∏î‡πâ ‡∏Å‡∏µ‡πà‡∏ä‡∏¥‡πâ‡∏ô (‡∏£‡∏ß‡∏° Quantity)


```python
count_product = df.groupby(['Region'])['Quantity'].sum()
count_product
```


 8. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡∏´‡∏≤‡∏Ñ‡πà‡∏≤ ‡∏¢‡∏≠‡∏î‡∏Ç‡∏≤‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏π‡∏á‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î (Max) ‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡πÅ‡∏ú‡∏ô‡∏Å Region


```python
max_region = df.groupby(['Region'])['Sales_Amount'].max()
max_region
```


‡∏´‡∏°‡∏ß‡∏î: ‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏ï‡∏≤‡∏£‡∏≤‡∏á (Sorting & Inspecting) 
9. ‡πÇ‡∏à‡∏ó‡∏¢‡πå: ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡∏≤‡∏° Sales_Amount ‡∏à‡∏≤‡∏Å ‡∏°‡∏≤‡∏Å‡πÑ‡∏õ‡∏ô‡πâ‡∏≠‡∏¢ 


```python
sorted_amount = df.sort_values('Sales_Amount', ascending=False)
sorted_amount 
```
